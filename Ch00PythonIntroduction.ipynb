### Arrays (NumPy)
* In numerical analysis, we often need fast and efficient operations on large sets of numbers.

* Python's NumPy library provides the array data structure, which is much more efficient than lists for numerical work.

* Arrays are homogeneous (all elements of the same type, usually numbers).

* They allow vectorized operations, meaning we can apply mathematical operations on the entire array at once (no need for explicit loops).

```python
a = np.array([1,2,3])
print(a*2)
```

```python
b = np.array([2,-3,5])
print(a*b)
```

```python
print(np.dot(a,b))
```

```python
print(np.cross(a,b))
```

```python
print(np.outer(a,b))
```

```python
A = np.array([[1,2,3],[-1,3,2],[2,-4,1]])
b = np.array([2,-3,5])
print(A*b)
```

```python
np.matmul(A,b)
```

```python
np.matmul(b,A)
```

```python
A = np.array([[1,2,3],[2,5,-1],[3,6,1]])
B = np.array([[1,0,0],[0,1,0],[0,0,1]])
np.matmul(A,B)
```

# Loops
Loops are programming structures that allow us to repeat a set of instructions multiple times without writing the same code again and again. Python mainly uses two types of loops: for loops and while loops.

### For Loop
A `for` loop is used when we know how many times we want to repeat something, or when we want to go through items in a sequence (like a list, string, or range of numbers).

It automatically takes each element one by one and executes the block of code.

```python
# Sum of first n natural numbers
s = 0
n = 5
for i in range(1,n+1):
    s += i

print("Sum = ", s)
```

### While Loop
A `while` loop is used when we don't know in advance how many times we need to repeat something.

Instead, it continues as long as a condition is true.

Care must be taken to avoid infinite loops (where the condition never becomes false).

```python
# Find the sum of non-negative numbers entered by user.
# Stop and display sum when a negative number is entered.

s = 0
x = float(input("Enter a number: "))
while x >= 0:
    s += x
    x = float(input("Enter next number: "))

print("\nSum = ", s)
```

### Exercise:
Input a postive integer `n` from the user and hence output the first `n` terms of a Fibonacci sequence.

```python
# Generate the Fibonacci sequence with given number of terms
def fibonacci(numTerms):
    # initialize
    a_1=1  # first term
    a_2=1  # second term

    print(a_1,", ",a_2, end="")

    for i in range(numTerms-2):
        # evaluate a term as sum of previous two terms
        a_n=a_1+a_2
        print(", ",a_n, end="")
        a_1=a_2
        a_2=a_n

n=input("Enter number of terms of Fibonacci sequence:")
print("Fibonacci :")
fibonacci(int(n))
```

```python
# Recursive method for generating terms of Fibonacci sequence
def fibr(n):
    if n==1 or n==2:
        return 1
    return fibr(n-1)+fibr(n-2)

for i in range(1,10):
    print(fibr(i),end = "  ")
```

# Plotting in python

```python
# @title {vertical-output:true}
import matplotlib.pyplot as plt

# define a vector for input variables for the graph
x=np.linspace(-2,2,100)
# define output variable for first function
y1=x**2
# define output variable for second function
y2=np.exp(x)

# define figure object with figure size
fig=plt.figure(figsize=[3,3])
ax = fig.add_subplot(1, 1, 1)
plt.xlabel("independent x")
plt.ylabel("dependent y")
plt.title("Example: 2D graph")

# creates the graph for the figure object
plt.plot(x,y1,'r',x,y2,'b--')
# display the graph
plt.show()
```

```python
# @title {vertical-output:true}
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 100

x = 10*np.random.rand(n)+10     # x in [10,20]
y = 20*np.random.rand(n)+5      # y in [5, 25]
z = 50*np.random.rand(n)-25     # z in [-25,25]
ax.scatter(x, y, z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# @title {vertical-output:true}
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

x = np.linspace(-1,1,50)
y = np.linspace(-1,1,50)
X, Y = np.meshgrid(x, y)
Z = np.exp(X**2+Y**2)
ax.plot_surface(X,Y,Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# @title {vertical-output:true}
import matplotlib.pyplot as plt

fig = plt.figure(figsize=[5,4])

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_colors = ['darkred', 'tab:blue', 'tab:red', 'tab:orange']

plt.bar(fruits, counts, label=fruits, color=bar_colors)

plt.ylabel('Fruit supply')
plt.title('Fruit supply by kind')

plt.show()
```

```python
# @title {vertical-output:true}
import matplotlib.pyplot as plt
import numpy as np

# define a vector for input variables for the graph
x = np.linspace(-2,2,100)
# define output variable for first function
y1 = np.exp(x)
# define output variable for second function
y2 = np.cos(x**2)

# define figure object with figure size
fig = plt.figure(figsize=[4,6])
ax = fig.add_subplot(2,1,1)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Intersecting graphs")
# creates the graph for the figure object
plt.plot(x,y1,'r',x,y2,'b--')
plt.legend(["exp(x)","cos(x^2)"])
# display the graph

x = np.linspace(0.5,2.5,100)
y3 = np.sin(1/x)
ax = fig.add_subplot(2, 1, 2)
plt.xlabel("x values")
plt.ylabel("y values")
plt.title("sin(1/x)")

plt.plot(x,y3)
plt.subplots_adjust(hspace=0.5)

x = np.linspace(1,2,50)
y4 = np.sin(1/x)
ax.fill_between(x,y4)
plt.show()
```
